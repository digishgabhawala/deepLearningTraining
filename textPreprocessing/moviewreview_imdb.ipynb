```python
import numpy as np
import keras 
import tensorflow as tf
```

```python
dataset = keras.datasets.imdb.load_data()
```

```python
word_indexes = keras.datasets.imdb.get_word_index()
```

```python
word_indexes
```

```python
index_to_word = {}
for key,val in word_indexes.items():
    index_to_word[val] = key
(x_train,y_train),(x_test,y_test) = dataset
```

```python
def get_line(index):
    return ' '.join([index_to_word[word] for word in x_train[index]])
print(get_line(0),y_train[0])
print(get_line(1),y_train[1])
print(get_line(2),y_train[2])
#print(get_line(18152),y_train[18152])
```

```python
#above seems like a bit cleaned dataset to use...
## also x data is quite like embedded dataset earlier 
```

```python
word_count = lambda sentence : len(sentence) 
longest_seq = max(x_train,key=word_count) 
longest_seq_length = len(longest_seq)
print(longest_seq,longest_seq_length)
```

```python
longest_seq_length=100#keeping max longest seq length to 100
padded_seq = keras.preprocessing.sequence.pad_sequences(x_train,longest_seq_length,padding="post")
print(padded_seq)
```

```python
vocab_length = len(index_to_word)+1

print(len(index_to_word),index_to_word[88584],word_indexes["'l'"])
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Embedding(vocab_length+2,100,input_length=longest_seq_length))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
print(x_train.shape)
start_seq = 15000
end_seq = 19000
padded_seq_data = padded_seq[start_seq:end_seq] 
y_train_data = y_train[start_seq:end_seq]
model.fit(padded_seq_data,y_train_data,epochs=1,verbose=1)
```

```python
model.fit(padded_seq,y_train,epochs=10,verbose=1,batch_size=500,shuffle=False)
```

```python
#prediction 
padded_test_seq = keras.preprocessing.sequence.pad_sequences(x_test,longest_seq_length,padding="post")
test_score = model.evaluate(padded_test_seq,y_test)
print(test_score)
```
