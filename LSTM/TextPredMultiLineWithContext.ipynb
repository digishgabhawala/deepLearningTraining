```python
import numpy as np
import tensorflow as tf
from tensorflow.contrib import rnn
import random
import collections
import os
import time
import pandas as pd
from keras.preprocessing.text import Tokenizer
import keras
```

```python
path_to_file = 'machinelearninglstm.txt'
with open(path_to_file) as f:
    text = f.readlines()
    
```

```python
text 
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)
dataset = tokenizer.texts_to_sequences(text)
```

```python
print(tokenizer.word_index)
print(dataset)
```

```python

def convert2DStrTo1DStr(dataset):
    out = []
    for x in dataset:
        for y in x:        
            out.append(y)
    return out
print(convert2DStrTo1DStr(dataset))
```

```python
encoded = convert2DStrTo1DStr(dataset)
```

```python
vocab_size = len(tokenizer.word_index)+1
print('Vocabulary Size: %d' % vocab_size)
```

```python
sequences = list()
seq_length = 3
x_data = []
y_data = []
for i in range(0, len(encoded)-seq_length):
    in_seq = encoded[i:i+seq_length]
    out_seq= encoded[i+seq_length]
    x_data.append(in_seq)
    y_data.append(out_seq)
```

```python
print('Total Sequences: %d' % len(y_data))
```

```python
print(x_data[0],y_data[0])
```

```python
X = np.array(x_data)
X = np.array(X).reshape(len(x_data),seq_length,1)
print(X.shape, X)
```

```python
Y = np.array(y_data)
Y_cat = keras.utils.to_categorical(Y)
Y_cat.shape
```

```python
model = keras.models.Sequential()
model.add(keras.layers.LSTM(200,input_shape=(seq_length,1)))
model.add(keras.layers.Dense(Y_cat.shape[1],activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X,Y_cat,epochs=500,verbose=2)
```

```python
def predict(current_words,tokenizer,model):
    encoded_input = np.array(tokenizer.texts_to_sequences([current_words])[0])
    if(len(encoded_input)>seq_length):
        encoded_input = encoded_input[-seq_length:]
    encoded_input = np.array(encoded_input).reshape(1,seq_length,1)#assuming current_words are only 3 words
    y_class = model.predict_classes(encoded_input)
    for word,index in tokenizer.word_index.items():
        if index==y_class:
            out_word = word
            break
    return word
def predict_n(current_words,tokenizer,model,n):
    next_words = current_words
    out = current_words
    for i in range(n):
        next_word = predict(out,tokenizer,model)
        out = out +' ' +  next_word
    return out
```
