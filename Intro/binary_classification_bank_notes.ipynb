```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.utils import shuffle
import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt',header=None)
```

```python
X = dataset.values[:,0:4]
Y= dataset.values[:,4]

```

```python
X,Y = shuffle(X,Y,random_state=1)

```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)

```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_shape=(X.shape[1],)))
model.add(Dense(1,activation='sigmoid'))

```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

```

```python
hisory = model.fit(X_train,Y_train,epochs=60,batch_size=128,verbose=1)

```

```python
score = model.evaluate(X_test,Y_test,batch_size=10)

```

```python
y_pred = model.predict(X_test)

```

```python
Y_test,y_pred

```

```python
sess = tf.Session()
rounded_y_pred = sess.run(tf.math.round(y_pred))
sess.close()
```
